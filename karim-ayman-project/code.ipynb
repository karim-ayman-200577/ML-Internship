### Dataset Column Descriptions

- Booking_ID: Unique identifier for each booking record.  
- number of adults: Number of adults included in the booking.  
- number of children: Number of children included in the booking.  
- number of weekend nights: Nights stayed during weekends (Friday–Saturday).  
- number of week nights: Nights stayed during weekdays (Sunday–Thursday).  
- type of meal: Meal plan chosen (e.g., Meal Plan 1, Meal Plan 2, or None).  
- car parking space: Indicates if a parking space was reserved (1 = yes, 0 = no).  
- room type: Type/category of room reserved.  
- lead time: Number of days between booking date and check-in date.  
- market segment type: Source of booking (e.g., Online, Offline, Corporate).  
- repeated: Whether the booking is from a returning customer (1 = yes, 0 = no).  
- P-C: Proxy column related to cancellations (possibly prediction target, 1 = canceled).  
- P-not-C: Proxy column for non-cancellations (possibly prediction target, 1 = not canceled).  
- average price: Average daily rate (ADR) paid per room.  
- special requests: Number of special requests made by the guest (e.g., extra bed, view).  
- date of reservation: Date when the booking was made.  
- booking status: Final status of the booking (Canceled / Not_Canceled).
